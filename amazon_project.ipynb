# Group By
#It is used to group the data based on columns in  a dataframe

```python
df.groupby(["Size"],as_index=False)["Qty"].sum().sort_values(by="Qty",ascending=False)
```

```python
df1=df.groupby(["Size"],as_index=False)["Qty"].sum().sort_values(by="Qty",ascending=False)
df2=sns.barplot(x="Size",y="Qty",data=df1)
for bars in df2.containers:
    df2.bar_label(bars)
```

# from above graph you can see most of the Qty buys M-size in the sales

# histogram

```python
df["Size"].hist()
```

```python
df["Category"] = df["Category"].astype(str)
column_data=df["Category"]
plt.figure(figsize=(10,5))
plt.hist(column_data,bins=30,edgecolor="red")
plt.xticks(rotation=90)
plt.show()           
```

# from above graph most of the buyers are T-shirts

```python
B2B_check=df["B2B"].value_counts()
plt.pie(B2B_check, labels=B2B_check.index,autopct='%1.1f%%')
plt.show()
```

# from above chart max of buyers are retailers and very less buyers are B2B buyers

```python
# prepare scatter plot
x_data=df["Category"]
y_data=df["Size"]

plt.scatter(x_data, y_data)
plt.xlabel("Category")
plt.ylabel("Size")
plt.title('scatter plot')
plt.show()
```

# from above chart you can see only free size is not available in T-shirts

```python
# Top 10 states
top_10_states=df["ship-state"].value_counts().head(10)
#plot count of cities by state
plt.figure(figsize=(10,5))
sns.countplot(data=df[df["ship-state"].isin(top_10_states.index)],x="ship-state")
plt.xlabel("ship-state")
plt.ylabel("count")
plt.title("distribution of state")
plt.xticks(rotation=45)
plt.show()
```

# from above charts most of the buyers are from maharashtra

# conclusion :=> The data analysis reveals that most of the customers are from maharashtra state,mainly serves retailers and fullfill
